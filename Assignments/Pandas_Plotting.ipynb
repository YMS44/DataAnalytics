```python
'''1. Write a Pandas program to create a line plot of the historical stock prices of
Alphabet Inc. between two specific dates.'''


df['Date'] = pd.to_datetime(df['Date'])
df['Date']

start_date = '2020-04-01'
end_date = '2020-09-30'

df_filtered = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

plt.plot(df_filtered['Date'], df_filtered['Close'])

plt.title('Historical Stock Prices of Alphabet Inc.')
plt.xlabel('Date')
plt.ylabel('Price')
```

```python
'''14. Write a Pandas program to create a plot of adjusted closing prices, thirty
days and forty days simple moving average of Alphabet Inc. between two specific
dates.'''

start_date ='2020-04-01'
end_date ='2020-04-30' # Adjust this as needed

filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

filtered_df['30-day SMA'] = filtered_df['Adj Close'].rolling(window=30).mean()
filtered_df['40-day SMA'] = filtered_df['Adj Close'].rolling(window=40).mean()


plt.figure(figsize=(10, 6))
plt.plot(filtered_df['Date'], filtered_df['Adj Close'], label='Adjusted Closing Price')
plt.plot(filtered_df['Date'], filtered_df['30-day SMA'], label='30-day SMA')
plt.plot(filtered_df['Date'], filtered_df['40-day SMA'], label='40-day SMA')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Adjusted Closing Prices and Moving Averages for Alphabet Inc.')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
'''15. Write a Pandas program to create a plot of adjusted closing prices, 30 days
simple moving average and exponential moving average of Alphabet Inc.
between two specific dates.'''

start_date ='2020-04-01'
end_date ='2020-04-30'
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

filtered_df['30-day SMA'] = filtered_df['Adj Close'].rolling(window=30).mean()


filtered_df['EMA'] = filtered_df['Adj Close'].ewm(span=30, adjust=False).mean()


plt.figure(figsize=(10, 6))
plt.plot(filtered_df['Date'], filtered_df['Adj Close'], label='Adjusted Closing Price')
plt.plot(filtered_df['Date'], filtered_df['30-day SMA'], label='30-day SMA')
plt.plot(filtered_df['Date'], filtered_df['EMA'], label='Exponential Moving Average (EMA)')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Adjusted Closing Prices, 30-day SMA, and EMA for Alphabet Inc.')
plt.legend()
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
'''16. Write a Pandas program to create a scatter plot of the trading volume/stock
prices of Alphabet Inc. stock between two specific dates.'''
start_date ='2020-04-01'
end_date ='2020-04-30'
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

filtered_df['Date'] = pd.to_datetime(filtered_df['Date'])

plt.figure(figsize=(15, 10))
plt.scatter(filtered_df['Close'], filtered_df['Volume'], s=100)
plt.grid(True)
plt.title('Trading Volume/Price of Alphabet Inc. stock\n01-04-2020 to 30-09-2020', fontsize=14, color='black')
plt.xlabel("Stock Price", fontsize=12, color='black')
plt.ylabel("Trading Volume", fontsize=12, color='black')
plt.show()
```

```python
'''17. Write a Pandas program to create a plot to visualize daily percentage returns
of Alphabet Inc. stock price between two specific dates.'''
start_date ='2020-04-01'
end_date ='2020-04-30'
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

filtered_df['Daily Returns'] = filtered_df['Adj Close'].pct_change() * 100

plt.figure(figsize=(10, 6))
plt.plot(filtered_df['Date'], filtered_df['Daily Returns'], color='b', marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Daily Percentage Returns')
plt.title('Daily Percentage Returns of Alphabet Inc. Stock Price\n01-04-2020 to 30-09-2020')
plt.grid(True)
plt.xticks(rotation=80)
plt.show()
```
