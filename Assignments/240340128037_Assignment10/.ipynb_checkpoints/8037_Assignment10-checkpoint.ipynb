```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_08 = pd.read_csv('all_alpha_08.csv')
df_18 = pd.read_csv('all_alpha_18.csv')
```

```python
# 1. Data Cleaning
# Drop unnecessary columns
df_08 = df_08.drop('Unadj Cmb MPG',axis=1)

# Handle missing values
df_08 = df_08.dropna()
df_18 = df_18.dropna()
```

```python
# 2. Inspection
print(df_08.head())
print(df_18.head())
print(df_08.describe())
print(df_18.describe())
print(df_08.info())
print(df_18.info())
```

```python
# 3. Manipulation if required 
# Convert fuel type to categorical
df_08['Fuel'] = df_08['Fuel'].astype('category')
df_18['Fuel'] = df_18['Fuel'].astype('category')


# Create a new column for combined mpg
df_08['Cmb MPG'] = pd.to_numeric(df_08['City MPG'],errors='coerce') * 0.55 + pd.to_numeric(df_08['Hwy MPG'],errors='coerce') * 0.45
df_18['Cmb MPG'] = pd.to_numeric(df_18['City MPG'],errors='coerce') * 0.55 + pd.to_numeric(df_18['Hwy MPG'],errors='coerce') * 0.45
```

```python
#4. Extracting meaningful visualization based on your understanding

plt.figure(figsize=(22, 12))
sns.scatterplot(x='City MPG', y='Hwy MPG', data=df_08, label='2008', alpha=0.8)
sns.scatterplot(x='City MPG', y='Hwy MPG', data=df_18, label='2018', alpha=0.4)
plt.xlabel('City MPG')
plt.ylabel('Highway MPG')
plt.title('City MPG vs. Highway MPG (2008 vs. 2018)')
plt.legend()
plt.show()
```

```python
#5. Data consolidation if required

# Merge the datasets using the 'Model' column as the key
merged_df = pd.merge(df_08, df_18, on='Model', suffixes=('_2008', '_2018'))


# Save the consolidated dataset to a new CSV file
merged_df.to_csv('fuel_economy_combined.csv', index=False)

print("Consolidated dataset saved as 'fuel_economy_combined.csv'")
```

```python
df_2008 = pd.read_csv('clean_08.csv')
df_2018 = pd.read_csv('clean_18.csv')
df_2018
```

```python
# Q1. Find out the alternative sources of fuel available in 2008 & 2018 respectively and by how much?
x=df_2008['fuel'].value_counts()
print(x)
```

```python
y=df_2018['fuel'].value_counts()
print(y)
```

```python
fuel_2008 = df_2008['fuel'].value_counts().sum()
fuel_2018 = df_2018['fuel'].value_counts().sum()

alternative_fuels_2008 = df_2008['fuel'].isin(['CNG', 'ethanol','gas']).value_counts()
alternative_fuels_2018 = df_2018['fuel'].isin(['CNG', 'Ethanol','Electricity','Gas']).value_counts()

print("Alternative sources of fuel in 2008:")
print((alternative_fuels_2008/fuel_2008)*100)

print("\nAlternative sources of fuel in 2018:")
print((alternative_fuels_2018/fuel_2018)*100)

import matplotlib.pyplot as plt

# Calculate the percentage of alternative fuels
alt_fuel_2008_pct = (alternative_fuels_2008 / fuel_2008) * 100
alt_fuel_2018_pct = (alternative_fuels_2018 / fuel_2018) * 100

# Create a dataframe for plotting
df_plot = pd.DataFrame({'2008': alt_fuel_2008_pct, '2018': alt_fuel_2018_pct})

# Plot the data
df_plot.plot.pie(subplots=True, figsize=(10, 5) , autopct = '%1.1f')

plt.show()
```

```python
alt_fuel_2008 = (x.sum()-x['Gasoline'])/x.sum()
alt_fuel_2018 = (y.sum()-y['Gasoline'])/y.sum()
print(alt_fuel_2008)
print(alt_fuel_2018)
print('Changes in alternative fuel from 2008 to 2018:{:.2f}%'.format((alt_fuel_2018-alt_fuel_2008)*100) )
```

```python
# Q2. Is there any improvement in ‘fuel economy’ with respect to vehicle class from 2008 to 2018, perform the necessary steps to examine?
```

```python

# Group by EPA Vehicle Class and calculate average combined mpg
avg_mpg_2008 = df_2008.groupby('veh_class')['cmb_mpg'].mean()
avg_mpg_2018 = df_2018.groupby('veh_class')['cmb_mpg'].mean()

# Calculate percentage improvement
percentage_improvement = ((avg_mpg_2018 - avg_mpg_2008) / avg_mpg_2008) * 100

# Create a bar chart to visualize the changes
plt.figure(figsize=(10, 6))
sns.barplot(x=percentage_improvement.index, y=percentage_improvement.values, palette='viridis')
plt.xlabel('Vehicle Class')
plt.ylabel('Percentage Improvement in Combined MPG')
plt.title('Fuel Economy Improvement by Vehicle Class (2008 to 2018)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show();
```

```python
# Q3. Is there any change in characteristics of SmartWay Vehicles ?
```

```python
# Filter data for SmartWay vehicles
smartway_2008 = df_2008[df_2008['smartway'] == 'yes']
smartway_2018 = df_2018[df_2018['smartway'] == 'Yes']

# Calculate average combined mpg for SmartWay vehicles
avg_mpg_2008 = smartway_2008['cmb_mpg'].mean()
avg_mpg_2018 = smartway_2018['cmb_mpg'].mean()

# Calculate percentage improvement in fuel economy
percentage_improvement = ((avg_mpg_2018 - avg_mpg_2008) / avg_mpg_2008) * 100

print(f"Average combined mpg for SmartWay vehicles (2008): {avg_mpg_2008:.2f}")
print(f"Average combined mpg for SmartWay vehicles (2018): {avg_mpg_2018:.2f}")
print(f"Percentage improvement: {percentage_improvement:.2f}%")

# Create a bar chart to visualize the changes
plt.figure(figsize=(8, 6))
sns.barplot(x=['2008', '2018'], y=[avg_mpg_2008, avg_mpg_2018], palette='viridis',width=0.25)
plt.xlabel('Year')
plt.ylabel('Average Combined MPG')
plt.title('Fuel Economy Improvement for SmartWay Vehicles (2008 to 2018)')
plt.show()
```

```python
# Q4. Which all features are associated with better fuel economy?
correlation = df_2018.select_dtypes(exclude=object).corr()
print(correlation)

cmap=sns.diverging_palette(10,220, as_cmap=True)
sns.heatmap(correlation, vmin=-1.0,annot=True,vmax=1.0, square=True , cmap=cmap)
plt.show()


```

```python
# Q5. How much improvement is there in miles/gallon or mpg? Also, which vehicle has improved the most?

# Calculate average mpg for each vehicle class in 2008
avg_mpg_2008 = df_2008.groupby('model')['cmb_mpg'].mean()

# Calculate average mpg for each vehicle class in 2018
avg_mpg_2018 = df_2018.groupby('model')['cmb_mpg'].mean()

# Calculate improvement in mpg
mpg_improvement = avg_mpg_2018 - avg_mpg_2008

# Find the vehicle class with the highest improvement
most_improved_class = mpg_improvement.idxmax()
most_improvement_value = mpg_improvement.max()

print(f"Average MPG Improvement by Vehicle Class (2008 to 2018):\n")
print(mpg_improvement)
print(f"\nMost Improved Vehicle Class: {most_improved_class} ({most_improvement_value:.2f} mpg improvement)")
```
