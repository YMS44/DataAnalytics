```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv("walmart_purchase_data.csv")
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

# Assuming you have a dataframe `df` with columns 'job' and 'company'
df.groupby('Job')['Company'].count().plot(kind='bar')
plt.title('Job vs Company')
plt.xlabel('Job')
plt.ylabel('Company')
plt.show()
```
