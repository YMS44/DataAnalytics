```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
```

```python
housing,target = fetch_california_housing(as_frame=True,return_X_y=True)
print(housing.head())
print(target.head())
```

```python
#Correlation of all the columns with each other
print(housing.corr())
```

```python
#Correlation of two specific column
corr  = housing.corr()
print(corr["MedInc"]["AveRooms"])
```

```python
#Visual View
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cmap = sns.diverging_palette(10,220,as_cmap=True)
sns.heatmap(corr,vmin=-1.0,vmax=1.0,square=True,cmap=cmap)
plt.show()
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df
```

```python
correlation_coefficient = df['reading score'].corr(df['writing score'])
print(f'Correlation Coefficient:{correlation_coefficient}')
```

```python
plt.scatter(df['reading score'],df['writing score'])
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.title(f'Scatter plot (Correlation: {correlation_coefficient:.2f})')
```

```python
residuals = df['writing score'] - df['reading score']
print(residuals)
```

```python
std_deviation = residuals.std()
```

```python
outliers = df[abs(residuals)>2*std_deviation]
plt.scatter(outliers['reading score'],outliers['writing score'],color='red',label='Outliers')
plt.legend()
plt.show()
```
