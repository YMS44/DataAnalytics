```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Administrator.DAI-PC2\Desktop\archive\Employee_monthly_salary.csv')
df
```

```python
d = df['Gender'].value_counts()
d
```

```python
from scipy import stats
# Separate the monthly income data into two groups: males and females
male_income = df[df['Gender'] == 'M']['Net_Pay']
female_income = df[df['Gender'] == 'F']['Net_Pay']

# Perform an independent t-test
t_stat, p_val = stats.ttest_ind(male_income, female_income)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")

# Interpret the p-value
alpha = 0.05  # significance level
if p_val < alpha:
    print("We reject the null hypothesis. There is a significant difference between the monthly incomes.")
else:
    print("We fail to reject the null hypothesis. There is no significant difference between the monthly incomes.")
```

```python
# Selecting 5 relevant columns for the histogram
columns_to_plot = ['Age', 'Tenure_in_org_in_months', 'GROSS', 'Net_Pay', 'Deduction']

# Plotting histograms
for column in columns_to_plot:
    plt.figure()
    plt.hist(df[column], bins=10, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
plt.figure(figsize=(10, 8))
df.boxplot(column=['Age', 'Tenure_in_org_in_months', 'GROSS', 'Net_Pay', 'Deduction'])
plt.title('Boxplot of Selected Columns')
plt.show()
```
