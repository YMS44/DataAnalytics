```python
import pandas as pd
data = pd.read_csv("salary.csv")
```

```python
#For X (independent variable take all the columns from the file,except salary)
X = data.iloc[:,:-1].values
```

```python
#For Y (dependent variable take all the second columns which is salary)

y = data.iloc[:,1].values
```

```python
#Step 2 : Split data into training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=1/3,random_state=0)
```

```python
#Step3 : Fit Simple Linear Regression to Training data
from sklearn.linear_model import LinearRegression
regressor =  LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#Step 4: Make Prediction
y_pred = regressor.predict(X_test)
print(y_pred)
```

```python
#Step 5: Visualize training set results
import matplotlib.pyplot as plt
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.title("Salary vs Experience(Training Set)")
plt.xlabel("Year of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
#Step 6 : Visualize test set results
plt.scatter(X_test,y_test,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.title("Salary vs Experience(Test Set)")
plt.xlabel("Year of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
#Step 7 : Make new prediction
new_salary_pred = regressor.predict([[15]])
print("The predicted salary of a person with 15 years experience is ",new_salary_pred)
```

```python
#Step 8: Intercept and coefficient
print("Intercept .. Fresher Salary : ",regressor.intercept_)
print("Coefficient .. Additional Salary for Each Addititonal Year's of Experience: ",regressor.coef_)
```

```python
#Evaluate results
from sklearn.metrics import mean_absolute_error,mean_squared_error
import numpy as np
```

```python
MAE  = mean_absolute_error(y_test,y_pred)
MSE = mean_squared_error(y_test,y_pred)
RMSE = np.sqrt(MSE)
```

```python
print(f'MAE: {MAE} , MSE: {MSE}, RMSE: {RMSE}')
```
